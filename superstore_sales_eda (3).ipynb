```python
import matplotlib.pyplot as plt
import seaborn as sns

# Question 1 - Category wise Sales
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='steelblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Question 2 - Region wise Profit
df.groupby('Region')['Profit'].sum().plot(kind='bar', color='green')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```

```python
# Question 3 - Ship Mode usage
df['Ship Mode'].value_counts().plot(kind='bar', color='orange')
plt.title('Most Used Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Question 4 - Monthly Sales Trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
df.groupby('Month')['Sales'].sum().plot(kind='line', color='purple')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Question 5 - Sub-Category wise Discount
df.groupby('Sub-Category')['Discount'].mean().sort_values(ascending=False).plot(kind='bar', color='red')
plt.title('Average Discount by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Average Discount')
plt.tight_layout()
plt.show()
```

```python
# Project Summary
print("=== Superstore Sales EDA - Key Insights ===")
print("1. Technology category has highest sales")
print("2. West region is most profitable")
print("3. Standard Class is most used shipping mode")
print("4. Sales peak in Nov-Dec every year (holiday season)")
print("5. Sales dip mid-year around July")
print("6. Binders have highest average discount")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('SampleSuperstore.csv', encoding='latin1')
```

```python
# Profit Margin by Category
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
df.groupby('Category')['Profit_Margin'].mean().plot(kind='bar', color='teal')
plt.title('Profit Margin by Category')
plt.ylabel('Profit Margin %')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Customers by Sales
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='barh', color='coral')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(), 
            annot=True, cmap='coolwarm')
plt.title('Correlation between Sales, Profit, Discount & Quantity')
plt.tight_layout()
plt.show()
```

```python
# Sales vs Profit Scatter Plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Sales', y='Profit', 
                hue='Category', size='Discount')
plt.title('Sales vs Profit by Category')
plt.tight_layout()
plt.show()
```

```python
# State wise Top 10 Sales
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='barh', color='purple')
plt.title('Top 10 City by Sales')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
```

```python
df['Year'].value_counts()
```

```python
df.groupby('Year')['Sales'].sum().plot(kind='bar', color='green')
plt.title('Year-wise Sales Growth')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Most Profitable Sub-Categories
df.groupby('Sub-Category')['Profit'].sum().sort_values().plot(kind='barh', color='orange')
plt.title('Profit by Sub-Category')
plt.xlabel('Total Profit')
plt.tight_layout()
plt.show()
```

```python
# Discount Impact on Profit
df['Discount_Range'] = pd.cut(df['Discount'], 
                               bins=[0, 0.1, 0.3, 0.5, 1.0],
                               labels=['Low','Medium','High','Very High'])
sns.boxplot(data=df, x='Discount_Range', y='Profit')
plt.title('Discount Impact on Profit')
plt.tight_layout()
plt.show()
```

```python
# Customer Segment Analysis
df.groupby('Segment')['Sales'].sum().plot(kind='pie', 
                                          autopct='%1.1f%%',
                                          colors=['steelblue','coral','green'])
plt.title('Sales by Customer Segment')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Cities by Profit
df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10).plot(kind='barh', color='teal')
plt.title('Top 10 Cities by Profit')
plt.xlabel('Total Profit')
plt.tight_layout()
plt.show()
```
